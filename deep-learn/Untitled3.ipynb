## Soal 3: Penjelasan `StandardScaler`

`StandardScaler` dari `sklearn.preprocessing` digunakan untuk menstandarisasi fitur dataset. Standardisasi (atau Z-score normalization) mentransformasi data sedemikian rupa sehingga memiliki rata-rata 0 dan standar deviasi 1. Rumusnya adalah:

$X_{scaled} = (X - \mu) / \sigma$

Di mana:
- $X$ adalah nilai fitur asli
- $\mu$ adalah rata-rata fitur
- $\sigma$ adalah standar deviasi fitur

**Mengapa perlu dijalankan?**

Standardisasi ini penting untuk banyak algoritma Machine Learning, terutama yang berbasis jarak (misalnya K-Nearest Neighbors, Support Vector Machines) atau yang sensitif terhadap skala fitur (misalnya Regresi Logistik, Neural Networks). Tanpa standardisasi, fitur dengan rentang nilai yang lebih besar akan mendominasi dan memengaruhi hasil model secara tidak proporsional, meskipun fitur dengan rentang nilai kecil mungkin sama pentingnya.

`fit_transform` melakukan dua hal:
1.  `fit`: Menghitung rata-rata (mean) dan standar deviasi (standard deviation) dari data training.
2.  `transform`: Menggunakan rata-rata dan standar deviasi yang dihitung untuk menstandarisasi data training.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Contoh data training
X_train = np.array([
    [100, 10],
    [150, 15],
    [200, 20],
    [50, 5],
    [125, 12]
])

print("Original X_train:\n", X_train)

# Inisialisasi StandardScaler
sc = StandardScaler()

# Menerapkan fit_transform pada X_train
X_train_scaled = sc.fit_transform(X_train)

print("\nScaled X_train:\n", X_train_scaled)
print("\nMean setelah scaling (seharusnya mendekati 0):", X_train_scaled.mean(axis=0))
print("Standard Deviation setelah scaling (seharusnya mendekati 1):", X_train_scaled.std(axis=0))
```

## Soal 4: Confusion Matrix, Akurasi, Presisi, dan Recall

Berdasarkan data yang diberikan:
- True Positive (TP): 42
- True Negative (TN): 83
- False Positive (FP): 27
- False Negative (FN): 27

### a. Tabel Confusion Matrix

Sebuah Confusion Matrix biasanya direpresentasikan sebagai berikut:

|                   | **Predicted Positive** | **Predicted Negative** |
| :---------------- | :--------------------- | :--------------------- |
| **Actual Positive**   | True Positive (TP)     | False Negative (FN)    |
| **Actual Negative** | False Positive (FP)    | True Negative (TN)     |

Dengan nilai yang diberikan:

|                   | **Predicted Positive** | **Predicted Negative** |
| :---------------- | :--------------------- | :--------------------- |
| **Actual Positive**   | 42                     | 27                     |
| **Actual Negative** | 27                     | 83                     |

```python
# Definisikan nilai TP, TN, FP, FN
TP = 42
TN = 83
FP = 27
FN = 27

print(f"TP: {TP}")
print(f"TN: {TN}")
print(f"FP: {FP}")
print(f"FN: {FN}")
```

### b. Hitung Akurasi (Accuracy)

Rumus Akurasi:

$Accuracy = (TP + TN) / (TP + TN + FP + FN)$

```python
# Hitung Akurasi
accuracy = (TP + TN) / (TP + TN + FP + FN)
print(f"Akurasi (Accuracy): {accuracy:.4f}")
```

### c. Hitung Presisi (Precision) dan Recall

Rumus Presisi:

$Precision = TP / (TP + FP)$

Rumus Recall:

$Recall = TP / (TP + FN)$

```python
# Hitung Presisi
precision = TP / (TP + FP)
print(f"Presisi (Precision): {precision:.4f}")

# Hitung Recall
recall = TP / (TP + FN)
print(f"Recall: {recall:.4f}")
```

## Soal 5: Hitung RMSE (Root Mean Squared Error)

RMSE adalah akar kuadrat dari rata-rata kuadrat dari error. Ini mengukur seberapa jauh prediksi model dari nilai aktual. Rumusnya adalah:

$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$

Di mana:
- $n$ adalah jumlah observasi
- $y_i$ adalah nilai aktual
- $\hat{y}_i$ adalah nilai prediksi

```python
from sklearn.metrics import mean_squared_error

# Data Fare Aktual dan Fare Prediksi
fare_actual = np.array([50, 10, 100, 500])
fare_predicted = np.array([45, 25, 100, 400])

print("Fare Aktual:", fare_actual)
print("Fare Prediksi:", fare_predicted)

# Hitung Mean Squared Error (MSE)
mse = mean_squared_error(fare_actual, fare_predicted)

# Hitung Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print(f"\nMean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
